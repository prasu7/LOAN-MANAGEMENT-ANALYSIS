```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sn
from pyspark.sql.functions import col, avg
import getpass


spark=SparkSession.builder.appName('demo').getOrCreate()
loan_data=spark.read.csv("Book1_dataset (1).csv",inferSchema=True,header=True)



def run_query(choice):
    
    #Query 1 : Find the total number of loans and the average credit score for each loan status
    if choice == 1:
        
        df = loan_data.groupBy("Loan_Status").agg(count("*").alias("Total_Loans"), avg("Credit_Score").alias("Avg_Credit_Score"))
        df.show()
        df.toPandas().plot(kind='bar', x='Loan_Status', y='Total_Loans', figsize=(6, 4))
        plot.title('Total Loans by Loan Status')
        plot.xlabel('Loan Status')
        plot.ylabel('Total Loans')
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
       
    
    #Query 2 : Find the top 10 customers with the highest annual income
    elif choice == 2:
        
        df = loan_data.select("Customer_ID", "Annual_Income").orderBy(desc("Annual_Income")).limit(10)
        df.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 3 : Find the loan status distribution for each home ownership type    
    elif choice == 3:
        
        df = loan_data.groupBy("Home_Ownership", "Loan_Status").agg(count("*").alias("Loan_Status_Count"))
        df.show()
        df_pd = df.toPandas()
        df_pd.pivot(index='Home_Ownership', columns='Loan_Status', values='Loan_Status_Count').plot(kind='bar', stacked=True)
        plot.title('Loan Status by Home Ownership')
        plot.xlabel('Home Ownership')
        plot.ylabel('Loan Status Count')
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 4 : Find the average current loan amount for each loan purpose    
    elif choice == 4:
        
        df = loan_data.groupBy("Purpose").agg(avg("Current_Loan_Amount").alias("Avg_Loan_Amount"))
        df.show()
        df_pd = df.toPandas()
        df_pd.plot(kind='bar', x='Purpose', y='Avg_Loan_Amount', rot=90)
        plot.title('Average Loan Amount by Purpose')
        plot.xlabel('Purpose')
        plot.ylabel('Average Loan Amount')
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 5 : Find the average monthly debt for customers with credit scores above 700    
    elif choice == 5:
        
        loan_data.filter(col("Credit_Score") > 700).agg(avg("Monthly_Debt").alias("Avg_Monthly_Debt")).show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 6 : Find the loan status distribution for each loan purpose    
    elif choice == 6:
        
        df = loan_data.groupBy("Purpose", "Loan_Status").agg(count("*").alias("Loan_Status_Count"))
        df.show()
        df_pd = df.toPandas()
        sn.set(style="whitegrid")
        sn.set(rc={'figure.figsize':(25,6)})
        sn.barplot(x="Purpose", y="Loan_Status_Count", hue="Loan_Status", data=df_pd)
        plot.title("Loan Status by Purpose")
        plot.xlabel("Purpose")
        plot.ylabel("Loan Status Count")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 7 : Find the average number of open accounts for customers with credit scores above 700    
    elif choice == 7:
        
        df = loan_data.filter(col("Credit_Score") > 700).groupBy("Loan_Status").agg(avg("Number_of_Open_Accounts").alias("Avg_Open_Accounts"))
        df.show()
        df_pd = df.toPandas()
        fig, ax = plot.subplots(figsize=(4,3))
        ax.bar(df_pd["Loan_Status"], df_pd["Avg_Open_Accounts"])
        ax.set_title("Average Number of Open Accounts by Loan Status for Customers with Credit Score > 700")
        ax.set_xlabel("Loan Status")
        ax.set_ylabel("Average Number of Open Accounts")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 8 : Find the average credit score for each home ownership type    
    elif choice == 8:
        
        df = loan_data.groupBy("Home_Ownership").agg(avg("Credit_Score").alias("Avg_Credit_Score"))
        df.show()
        df_pd = df.toPandas()
        sn.set_style("whitegrid")
        plot.figure(figsize=(6, 3))
        sn.barplot(x="Home_Ownership", y="Avg_Credit_Score", data=df_pd)
        plot.title("Average Credit Score by Home Ownership")
        plot.xlabel("Home Ownership")
        plot.ylabel("Average Credit Score")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 9 : Find the average current credit balance for each loan purpose    
    elif choice == 9:
        
        df = loan_data.groupBy("Purpose").agg(avg("Current_Credit_Balance").alias("Avg_Credit_Balance"))
        df.show()
        df_pd = df.toPandas()
        sn.set_style("whitegrid")
        plot.figure(figsize=(20, 6))
        sn.barplot(x="Purpose", y="Avg_Credit_Balance", data=df_pd)
        plot.title("Average Current Credit Balance by Purpose")
        plot.xlabel("Purpose")
        plot.ylabel("Average Current Credit Balance")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
        
    #Query 10 : Find the total number of loans and the average annual income for each loan status    
    elif choice == 10:
        
        df = loan_data.groupBy("Purpose").agg(avg("Current_Credit_Balance").alias("Avg_Credit_Balance"))
        df.show()
        df_pd = df.toPandas()
        sn.set_style("whitegrid")
        plot.figure(figsize=(20, 6))
        sn.barplot(x="Purpose", y="Avg_Credit_Balance", data=df_pd)
        plot.title("Average Current Credit Balance by Purpose")
        plot.xlabel("Purpose")
        plot.ylabel("Average Current Credit Balance")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
            
        
    # Query 11: Find the top 10 customers with the highest monthly debt
    elif choice == 11:
        loan_data.select("Customer_ID", "Monthly_Debt").orderBy(desc("Monthly_Debt")).limit(10).show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        

    # Query 12: Find the loan status distribution for each term (Short Term or Long Term)
    elif choice == 12:
        
        df = loan_data.groupBy("Term", "Loan_Status").agg(count("*").alias("Loan_Status_Count"))
        df.show()
        df_pd = df.toPandas()
        sn.set_style("whitegrid")
        plot.figure(figsize=(20, 6))
        sn.barplot(x="Loan_Status", y="Loan_Status_Count", data=df_pd)
        plot.title("loan status distribution for each term (Short Term or Long Term)")
        plot.xlabel("Loan_Status")
        plot.ylabel("Loan_Status_Count")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        

    # Query 13: Find the average number of open accounts for customers with a current loan amount greater than $500,000
    elif choice == 13:
        
        loan_data.filter(col("Current_Loan_Amount") > 500000).agg(avg("Number_of_Open_Accounts").alias("Avg_Open_Accounts")).show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
    # Query 14: Find the average credit score for each loan purpose and home ownership type combination
    elif choice == 14:
        
        df = loan_data.groupBy("Purpose", "Home_Ownership").agg(avg("Credit_Score").alias("Avg_Credit_Score"))
        df.show()
        df_pd = df.toPandas()
        sn.set_style("whitegrid")
        plot.figure(figsize=(20, 6))
        sn.barplot(x="Purpose", y="Avg_Credit_Score", data=df_pd)
        plot.title("average credit score for each loan purpose and home ownership type combination")
        plot.xlabel("Purpose")
        plot.ylabel("Avg_Credit_Score")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
        
    
    # Query 15: Find the loan status distribution for each number of open accounts range (0-5, 6-10, 11-15, 16-20, 21-25, 26+)
    elif choice == 15:
        
        df = loan_data.withColumn("Open_Accounts_Range", when(col("Number_of_Open_Accounts") < 6, "0-5")
                    .when((col("Number_of_Open_Accounts") >= 6) & (col("Number_of_Open_Accounts") <= 10), "6-10")
                    .when((col("Number_of_Open_Accounts") >= 11) & (col("Number_of_Open_Accounts") <= 15), "11-15")
                    .when((col("Number_of_Open_Accounts") >= 16) & (col("Number_of_Open_Accounts") <= 20), "16-20")
                    .when((col("Number_of_Open_Accounts") >= 21) & (col("Number_of_Open_Accounts") <= 25), "21-25")
                    .otherwise("26+")).groupBy("Open_Accounts_Range", "Loan_Status").agg(count("*").alias("Loan_Status_Count"))
        df.show()
        df_pd = df.toPandas()
        sn.set_style("whitegrid")
        plot.figure(figsize=(20, 6))
        sn.barplot(x="Open_Accounts_Range", y="Loan_Status_Count", data=df_pd)
        plot.title("loan status distribution for each number of open accounts range (0-5, 6-10, 11-15, 16-20, 21-25, 26+)")
        plot.xlabel("Open_Accounts_Range")
        plot.ylabel("Loan_Status_Count")
        plot.show()
        tc=input("If you want to continue press y : ").lower()
        if tc=="y":
            run()
            
    
    elif choice == 16:
        print("------------- _/\_ Thank You Visit Again _/\_ -------------")

    else:
        print("Invalid choice")

    
    
#while True:
def run():
    print('''
            1. Find the total number of loans and the average credit score for each loan status
            2. Find the top 10 customers with the highest annual income
            3. Find the loan status distribution for each home ownership type
            4. Find the average current loan amount for each loan purpose
            5. Find the average monthly debt for customers with credit scores above 700
            6. Find the loan status distribution for each loan purpose
            7. Find the average number of open accounts for customers with credit scores above 700
            8. Find the average credit score for each home ownership type
            9. Find the average current credit balance for each loan purpose
            10. Find the total number of loans and the average annual income for each loan status
            11. Find the top 10 customers with the highest monthly debt
            12. Find the loan status distribution for each term (Short Term or Long Term)
            13. Find the average number of open accounts for customers with a current loan amount greater than $500,000
            14. Find the average credit score for each loan purpose and home ownership type combination
            15. Find the loan status distribution for each number of open accounts range (0-5, 6-10, 11-15, 16-20, 21-25, 26+)
            16. Exit   ''')

    choice = int(input("Enter your choice: "))
    
    run_query(choice)
    
username = "Loan@123"
pswd = "1122"

def login():
    a=input("enter the user name: ")
    b=getpass.getpass(prompt='Enter your password: ')
    if username==a and pswd==b:
        print("Logged in Successfully")
        run()
    else:
        print("Invalid User Name Or Password ")
        login()

login()
```
